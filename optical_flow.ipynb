```python
import numpy as np
import cv2
import tensorflow as tf

currentFrame=0
```

```python
#Path
source="E:/projects/Ball speed detection/dataset/own video data/baseball_23.Mp4"
Destination="E:/projects/Ball speed detection/dataset/own video data/images/"
```

```python



cap = cv2.VideoCapture(source)

while(True):
    # Capture frame-by-frame
    ret, frame = cap.read()

    # Saves image of the current frame in jpg file
    name = Destination + str(currentFrame) + '.jpg'
    print ('Creating...' + name)
    cv2.imwrite(name, frame)

    # To stop duplicate images
    currentFrame += 1

# When everything done, release the capture
cap.release()
cv2.destroyAllWindows()
```
